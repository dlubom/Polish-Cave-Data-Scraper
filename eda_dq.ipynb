```python
import pandas as pd

pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)
pd.set_option("display.width", 100)
pd.set_option("max_colwidth", 100)

import pyspark.sql.functions as F
import pyspark.sql.types as T
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("EDA_DQ_Caves").getOrCreate()

df = spark.read.parquet("caves_transformed.parquet")

df.createOrReplaceTempView("caves_view")
```

```python
spark.sql("SELECT * FROM caves_view WHERE latitude IS NULL or longitude IS NULL").toPandas()
```

```python
spark.sql("SELECT * FROM caves_view WHERE destroyed_unavailable_or_unfound NOT IN ('Tak', 'Nie')").toPandas()
```

```python
spark.sql("SELECT state_in_year, count(1) as cnt FROM caves_view GROUP BY state_in_year ORDER BY state_in_year DESC").toPandas()
```

```python
spark.sql("SELECT min(latitude), max(latitude), min(longitude), max(longitude) FROM caves_view").toPandas()
```

```python
spark.sql("SELECT destroyed_unavailable_or_unfound, count(1) as cnt FROM caves_view GROUP BY destroyed_unavailable_or_unfound ORDER BY cnt DESC").toPandas()
```

```python
df.select(F.explode("images").alias("image")).limit(10).toPandas()
```

```python
#  |-- images: array (nullable = true)
#  |    |-- element: struct (containsNull = true)
#  |    |    |-- metadata: struct (nullable = true)
#  |    |    |    |-- graphics_id: long (nullable = true)
#  |    |    |    |-- mime_type: string (nullable = true)
#  |    |    |    |-- creation_date: string (nullable = true)
#  |    |    |    |-- creation_date_string: string (nullable = true)
#  |    |    |    |-- graphics_name: string (nullable = true)
#  |    |    |    |-- author_name: string (nullable = true)
#  |    |    |    |-- graphics_type_id: long (nullable = true)
#  |    |    |    |-- graphics_type_name: string (nullable = true)
#  |    |    |    |-- inventory_number: string (nullable = true)
#  |    |    |    |-- maxWidth: long (nullable = true)
#  |    |    |    |-- maxHeight: long (nullable = true)

df_images = df.select(F.explode("images").alias("image"))

df_images = df_images.select(
    F.col("image.metadata.graphics_id").alias("graphics_id"),
    F.col("image.metadata.mime_type").alias("mime_type"),
    F.col("image.metadata.creation_date").alias("creation_date"),
    F.col("image.metadata.creation_date_string").alias("creation_date_string"),
    F.col("image.metadata.graphics_name").alias("graphics_name"),
    F.col("image.metadata.author_name").alias("author_name"),
    F.col("image.metadata.graphics_type_id").alias("graphics_type_id"),
    F.col("image.metadata.graphics_type_name").alias("graphics_type_name"),
    F.col("image.metadata.inventory_number").alias("inventory_number"),
    F.col("image.metadata.maxWidth").alias("max_width"),
    F.col("image.metadata.maxHeight").alias("max_height"),
    F.col("image.image_path").alias("image_path"),
    F.col("image.description").alias("image_description")
)
```

```python
df_images.printSchema()
```

```python
df_images.limit(10).toPandas()
```

```python
df_images.createOrReplaceTempView("images_view")
```

```python
spark.sql("SELECT mime_type, count(1) as cnt FROM images_view GROUP BY mime_type ORDER BY cnt DESC").toPandas()
```
